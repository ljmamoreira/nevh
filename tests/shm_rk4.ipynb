```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rhs(xv, w2=1.0):
    x, v = xv
    return np.array([v, -w2*x])
```

```python
xv_p = np.array([1., 0.])
dt = 0.452
w2 = 4*np.pi**2
tmin=0; tmax=25
n_iters = int((tmax-tmin)/dt)
trajectory = np.zeros((n_iters, 3))
trajectory[:,0] = np.arange(n_iters)*dt
trajectory[0,1:] = xv_p
for i in range(1, n_iters):
    k1 = rhs(xv_p, w2)
    k2 = rhs(xv_p + dt * k1 / 2, w2)
    k3 = rhs(xv_p + dt * k2 / 2, w2)
    k4 = rhs(xv_p + dt * k3, w2)
    xv_n = xv_p + (k1 + 2*(k2+k3) + k4) * dt / 6
    trajectory[i,1:] = xv_n
    xv_p = xv_n
    
```

```python
plt.ylim(-100,100)
plt.plot(trajectory[:,0], trajectory[:,1], trajectory[:,0], trajectory[:,2]);
```
